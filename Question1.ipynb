```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading the data into a dataframe
data =pd.read_csv('/bin/healthcare_dataset.csv')
print(data)
```

```python
# Question 1: Create a plot to find out if there is any correlation between Blood Group Type, Gender, and Medical Condition
plt.figure(figsize=(10, 6))
data.groupby(['Blood Group Type', 'Gender', 'Medical Condition']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Correlation between Blood Group Type, Gender, and Medical Condition')
plt.xlabel('Blood Group Type and Gender')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Medical Condition')
plt.tight_layout()
plt.show()
```

```python
# Question 2: Find out the Average Billing Amount required to treat each medical condition
average_billing_amount = data.groupby('Medical Condition')['Billing Amount'].mean()
print("Average Billing Amount required to treat each medical condition:")
print(average_billing_amount)
```
